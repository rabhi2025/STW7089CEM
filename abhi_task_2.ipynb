```r
# Load your dataset
data <- read.csv("dataset.csv")

# Model 1: y = θ1*x4 + θ2*(x3)^2 + θbias
model1 <- lm(x2 ~ x4 + I(x3^2), data = data)
summary(model1)
coef(model1)

# Model 2: y = θ1*x4 + θ2*(x3)^2 + θ3*x5 + θbias
model2 <- lm(x2 ~ x4 + I(x3^2) + x5, data = data)
summary(model2)
coef(model2)

# Model 3: y = θ1*x3 + θ2*x4 + θ3*(x5)^3
model3 <- lm(x2 ~ x3 + x4 + I(x5^3), data = data)
summary(model3)
coef(model3)

# Model 4: y = θ1*x4 + θ2*(x3)^2 + θ3*(x5)^2 + θbias
model4 <- lm(x2 ~ x4 + I(x3^2) + I(x5^2), data = data)
summary(model4)
coef(model4)

# Model 5: y = θ1*x4 + θ2*(x1)^2 + θ3*(x3)^2 + θbias
model5 <- lm(x2 ~ x4 + I(x1^2) + I(x3^2), data = data)
summary(model5)
coef(model5)
```

```r
# Model 1
rss1 <- sum(residuals(model1)^2)

# Model 2
rss2 <- sum(residuals(model2)^2)

# Model 3
rss3 <- sum(residuals(model3)^2)

# Model 4
rss4 <- sum(residuals(model4)^2)

# Model 5
rss5 <- sum(residuals(model5)^2)

# Print all RSS values
rss_values <- data.frame(
  Model = paste("Model", 1:5),
  RSS = c(rss1, rss2, rss3, rss4, rss5)
)
print(rss_values)
```

```r
# Number of observations
n <- nrow(data)

# Compute log-likelihood manually for each model
logLik_model1 <- -n/2 * log(2*pi) - n/2 * log(rss1/n) - rss1/(2 * (rss1/n))
logLik_model2 <- -n/2 * log(2*pi) - n/2 * log(rss2/n) - rss2/(2 * (rss2/n))
logLik_model3 <- -n/2 * log(2*pi) - n/2 * log(rss3/n) - rss3/(2 * (rss3/n))
logLik_model4 <- -n/2 * log(2*pi) - n/2 * log(rss4/n) - rss4/(2 * (rss4/n))
logLik_model5 <- -n/2 * log(2*pi) - n/2 * log(rss5/n) - rss5/(2 * (rss5/n))

# Display results
logLik_values <- data.frame(
  Model = paste("Model", 1:5),
  LogLikelihood = c(logLik_model1, logLik_model2, logLik_model3, logLik_model4, logLik_model5)
)

print(logLik_values)
```

```r
# Define number of observations
n <- nrow(data)

# Define number of parameters for each model (including intercept)
k1 <- 3  # Model 1: x4, x3^2, intercept
k2 <- 4  # Model 2: x4, x3^2, x5, intercept
k3 <- 4  # Model 3: x3, x4, x5^3, intercept
k4 <- 4  # Model 4: x4, x3^2, x5^2, intercept
k5 <- 4  # Model 5: x4, x1^2, x3^2, intercept

# Use previously computed log-likelihoods
logL <- c(logLik_model1, logLik_model2, logLik_model3, logLik_model4, logLik_model5)
k <- c(k1, k2, k3, k4, k5)

# Compute AIC and BIC
aic <- 2 * k - 2 * logL
bic <- k * log(n) - 2 * logL

# Compile results
info_criteria <- data.frame(
  Model = paste("Model", 1:5),
  AIC = round(aic, 2),
  BIC = round(bic, 2)
)

print(info_criteria)
```

```r
# Load data
data <- read.csv("dataset.csv")

# Add squared features used in Model 5
data$x1_sq <- data$x1^2
data$x3_sq <- data$x3^2

# Split dataset: 70% train, 30% test
set.seed(123)  # for reproducibility
sample_size <- floor(0.7 * nrow(data))
train_indices <- sample(seq_len(nrow(data)), size = sample_size)

train_data <- data[train_indices, ]
test_data <- data[-train_indices, ]

# Fit Model 5 to training data
model5_train <- lm(x2 ~ x4 + x1_sq + x3_sq, data = train_data)

# Predict on test data with 95% confidence intervals
predictions <- predict(model5_train, newdata = test_data, interval = "confidence", level = 0.95)

# Combine predictions and actual values
results <- data.frame(
  Actual = test_data$x2,
  Predicted = predictions[, "fit"],
  Lower95 = predictions[, "lwr"],
  Upper95 = predictions[, "upr"]
)

# Add an index column to track sample number
results$Index <- 1:nrow(results)

# Reshape data for plotting multiple lines
library(tidyr)
library(dplyr)

# Prepare data for ggplot
plot_data <- results %>%
  select(Index, Actual, Predicted) %>%
  pivot_longer(cols = c("Actual", "Predicted"),
               names_to = "Type", values_to = "Value")

# Plot with legend
library(ggplot2)

ggplot(plot_data, aes(x = Index, y = Value, color = Type)) +
  geom_point(data = subset(plot_data, Type == "Actual"), size = 1.5, alpha = 0.6) +
  geom_line(data = subset(plot_data, Type == "Predicted"), linewidth = 1) +
  geom_ribbon(data = results,
              aes(x = Index, ymin = Lower95, ymax = Upper95),
              inherit.aes = FALSE,
              fill = "lightblue", alpha = 0.4, show.legend = TRUE) +
  scale_color_manual(values = c("Actual" = "black", "Predicted" = "blue")) +
  labs(title = "Model 5 Predictions with 95% Confidence Intervals",
       x = "Test Sample Index",
       y = "Net Hourly Energy Output (x2)",
       color = "Legend") +
  theme_minimal()
```

```r
# Load necessary libraries
library(ggplot2)

# Load and prepare data
data <- read.csv("dataset.csv")
data$x1_sq <- data$x1^2
data$x3_sq <- data$x3^2

# Fixed parameters (from Task 2.1 estimates)
theta_fixed <- list(
  intercept = 251.7,
  x3_sq = -0.003651
)

# Prior ranges for the two parameters
prior_x4 <- c(0.1136, 0.3408)      # θ1: x4
prior_x1_sq <- c(-0.05478, -0.01826)  # θ2: x1^2

# Number of samples and threshold
n_samples <- 10000
epsilon <- 5000  # Adjust threshold for acceptance as needed

# Sample storage for accepted parameters
accepted_samples <- data.frame(x4 = numeric(), x1_sq = numeric())

set.seed(123)  # For reproducibility

# Subset data for faster computation (optional)
abc_data <- data[sample(1:nrow(data), 500), ]

for (i in 1:n_samples) {
  # Sample parameters from uniform priors
  theta_x4 <- runif(1, prior_x4[1], prior_x4[2])
  theta_x1_sq <- runif(1, prior_x1_sq[1], prior_x1_sq[2])

  # Calculate predicted y
  y_pred <- theta_fixed$intercept +
            theta_x4 * abc_data$x4 +
            theta_x1_sq * abc_data$x1_sq +
            theta_fixed$x3_sq * abc_data$x3_sq

  # Compute RSS
  rss <- sum((abc_data$x2 - y_pred)^2)

  # Accept if RSS < epsilon
  if (rss < epsilon) {
    accepted_samples <- rbind(accepted_samples, data.frame(x4 = theta_x4, x1_sq = theta_x1_sq))
  }
}

cat("Accepted samples:", nrow(accepted_samples), "\n")


```

```r
epsilon <- 20000  # increased threshold

prior_x4 <- c(0, 0.5)         # wider prior range
prior_x1_sq <- c(-0.08, 0)    # wider prior range

abc_data <- data[sample(1:nrow(data), 200), ]  # smaller subset for speed

accepted_samples <- data.frame(x4 = numeric(), x1_sq = numeric())

for (i in 1:n_samples) {
  theta_x4 <- runif(1, prior_x4[1], prior_x4[2])
  theta_x1_sq <- runif(1, prior_x1_sq[1], prior_x1_sq[2])

  y_pred <- theta_fixed$intercept +
            theta_x4 * abc_data$x4 +
            theta_x1_sq * abc_data$x1_sq +
            theta_fixed$x3_sq * abc_data$x3_sq

  rss <- sum((abc_data$x2 - y_pred)^2)

  if (rss < epsilon) {
    accepted_samples <- rbind(accepted_samples, data.frame(x4 = theta_x4, x1_sq = theta_x1_sq))
  }
}

cat("Accepted samples:", nrow(accepted_samples), "\n")
```
